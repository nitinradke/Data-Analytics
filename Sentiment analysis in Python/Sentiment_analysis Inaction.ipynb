```python

class listner(StreamListener):
    positive=0
    negative=0
    def on_data(self,data):
        #Loading Data
        all_data = json.loads(data)
        
        
        # Tweets to String
        tweet =all_data["text"]
        print(tweet)
        
#         Writing them on to file
        file = open("tweet_data","a")
        file.write(tweet)
        file.close()
        
        #Getting the status of tweet
        status = s.Sentiments(tweet)
        print(status)
        print("\n\n")
        
        #Counting posiitive or negative tweet
        if(status == 'pos'):
            listner.positive+=1
        else:
            listner.negative+=1
            
            
    def on_error(self,status):
        print(status)
        
```

```python
twitterstream = Stream(auth,listner())
```

```python
d = {'Positive_tweet':listner.positive,'Negative':listner.negative}
```

```python
modi_data = pd.DataFrame(d,index=range(0,1))
```
